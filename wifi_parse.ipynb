```python
import pandas as pd
import os
```

```python
path = 'wireless/'
```

```python
df_wifi = pd.DataFrame()
```

```python
for file in  os.listdir(path):
    d = pd.read_csv(path+file,delimiter='|')
    d['Station'] = file.replace('.csv','').replace('_',' ').capitalize()
    df_wifi = pd.concat([df_wifi,d],ignore_index=True,)
```

```python
df_wifi.to_csv('datasets/global_wifi.csv',index=False)
```
